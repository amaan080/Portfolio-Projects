# Data Cleaning

```python
# Removing unnamed Column
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
# Check & Fill null values
df.isna().sum()
```

```python
df["type"].value_counts()
```

```python
# Check for inconsistencies in categorical variables
print(df['type'].value_counts().sort_index())
```

```python
# Remove duplicates
df = df.drop_duplicates()
print(df)
```

# Descriptive Statistics

```python
# Calculate median
median_price = df['AveragePrice'].median()
print(median_price)

print()

# Calculate mode
mode_region = df['region'].mode()
print(mode_region)


print()

# Calculate variance
variance_total_volume = df['Total Volume'].var()
print(variance_total_volume)


print()

# Calculate standard deviation
std_deviation_4046 = df['4046'].std()
print(std_deviation_4046)
```

 # Histograms for Distribution Average price

```python

plt.figure(figsize=(12, 6))
df['AveragePrice'].hist(bins=30, color='blue')
plt.title('Distribution of Average Price')
plt.xlabel('Average Price')
plt.ylabel('Frequency')
plt.show()
```

# Box plots for Average Price by Type

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='type', y='AveragePrice', data=df)
plt.title('Average Price by Type')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.show()
```

# Create a pair plot to visualize relationships between multiple variables

```python
sns.pairplot(df, vars=["AveragePrice", "Total Volume", "Total Bags"])
plt.show()
```

# Correlation Matrix

```python
# Selecting only numerical columns for correlation analysis
numeric_data = df.select_dtypes(include='number')

# Calculate correlation matrix for numerical columns
correlation_matrix = numeric_data.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Scatter plot

```python

plt.scatter(df['Total Volume'], df['AveragePrice'])
plt.title('Total Volume vs Average Price')
plt.xlabel('Total Volume')
plt.ylabel('Average Price')
plt.show()
```

# Bar plot for Type

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

# Time series plot for 'Total Volume' over time

```python
# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Total Volume'])
plt.xlabel('Date')
plt.ylabel('Total Volume')
plt.title('Total Volume Over Time')
plt.show()
```
